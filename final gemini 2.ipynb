```python
def obtener_serie_usuario():
    """
    Pregunta al usuario por su serie favorita y la devuelve en formato de texto.
    """
    serie_favorita = input("📺 ¿Cuál es tu serie favorita? ")
    return serie_favorita
```

```python
def generar_recomendaciones(serie_favorita):
    """
    Utiliza Google Gemini para generar recomendaciones de series similares
    basadas en la serie favorita ingresada por el usuario.
    """
    prompt = f"""
    Eres un experto en series de televisión y streaming. 
    Un usuario te dice que su serie favorita es "{serie_favorita}". 
    Recomiéndale al menos 5 series similares y explica brevemente por qué podrían gustarle.
    """
    
    model = genai.GenerativeModel("gemini-pro")
    respuesta = model.generate_content(prompt)
    
    return respuesta.text  # Devuelve solo el texto generado
```

```python
def extraer_nombres_series(texto_recomendaciones):
    """
    Utiliza Google Gemini para extraer los nombres de las series recomendadas 
    a partir del texto generado en la recomendación.
    """
    prompt = f"""
    Extrae solo los nombres de las series mencionadas en el siguiente texto: 
    {texto_recomendaciones}
    
    Devuelve únicamente una lista separada por comas sin explicaciones adicionales.
    """
    
    model = genai.GenerativeModel("gemini-pro")
    respuesta = model.generate_content(prompt)
    
    return respuesta.text  # Devuelve solo los nombres de las series
```

```python
def generar_imagen_series(nombres_series):
    """
    Usa Google Gemini para generar una imagen representativa de las series recomendadas.
    """
    prompt = f"""
    Genera una imagen representativa que combine elementos visuales de las siguientes series de televisión: 
    {nombres_series}.
    
    La imagen debe reflejar los estilos y temáticas de estas series.
    """
    
    model = genai.GenerativeModel("gemini-pro-vision")
    respuesta = model.generate_content(prompt)
    
    return respuesta.text  # Devuelve la descripción de la imagen generada
```

```python
def recomendador_de_series():
    """
    Ejecuta todo el flujo del recomendador de series:
    1. Solicita la serie favorita al usuario.
    2. Obtiene recomendaciones de Google Gemini.
    3. Extrae los nombres de las series recomendadas.
    4. Genera una imagen representativa de las recomendaciones.
    5. Muestra los resultados al usuario.
    """
    
    # Paso 1: Pedir serie favorita al usuario
    serie_favorita = obtener_serie_usuario()
    
    # Paso 2: Generar recomendaciones
    texto_recomendaciones = generar_recomendaciones(serie_favorita)
    print("\n🎬 Recomendaciones de series similares:\n")
    print(texto_recomendaciones)
    
    # Paso 3: Extraer solo los nombres de las series
    nombres_series = extraer_nombres_series(texto_recomendaciones)
    print("\n📜 Series recomendadas (solo nombres):\n")
    print(nombres_series)
    
    # Paso 4: Generar imagen representativa
    imagen_generada = generar_imagen_series(nombres_series)
    print("\n🖼️ Descripción de la imagen generada:\n")
    print(imagen_generada)
    
# Ejecutar el recomendador
recomendador_de_series()
```
